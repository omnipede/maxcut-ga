```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
```

```python
fname="values.txt"
POP_SIZE=100
values=np.loadtxt(fname,delimiter=" ",skiprows=POP_SIZE*0,max_rows=POP_SIZE*1,dtype=int)
values=values.tolist()

x = [i[0] for i in values]
y = [i[1] for i in values]

df = pd.DataFrame({"generation": x,
                   "values": y})
sns.lmplot(x= 'generation', y = 'values', data = df)
```

```python
values=np.loadtxt(fname,delimiter=" ",skiprows=POP_SIZE*1,max_rows=POP_SIZE*5,dtype=int)
values=values.tolist()

x = [i[0] for i in values]
y = [i[1] for i in values]

df = pd.DataFrame({"generation": x,
                   "values": y})
sns.lmplot(x= 'generation', y = 'values', data = df)
```
